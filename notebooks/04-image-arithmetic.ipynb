Import Packages

```python
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import argparse
import cv2
```

Define display function

```python
def plt_imshow(title, image):
    # convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

Argument parser

```python
args = {
    "image": "../images/output/uss-freedom-rotate-90cc.jpg"
}
```

Load the image

```python
# load the image and show it
image = cv2.imread(args["image"])
plt_imshow("Original", image)
```

Lighten the image

```python
M = np.ones(image.shape, dtype="uint8") * 100
added = cv2.add(image, M)
plt_imshow("Lighter", added)
cv2.imwrite("../images/output/uss-freedom-rotate-90cc-lt.jpg", added)
```

Darken the image

```python
M = np.ones(image.shape, dtype="uint8") * 70
subtracted = cv2.subtract(image, M)
plt_imshow("Darker", subtracted)
cv2.imwrite("../images/output/uss-freedom-rotate-90cc-dk.jpg", subtracted)
```
