Import Packages

```python
# import the necessary packages
import matplotlib.pyplot as plt
import argparse
import imutils
import cv2
```

Define display function

```python
def plt_imshow(title, image):
    # convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

Argument parse

```python
args = {
    "image": "../images/output/uss-freedom-resized.jpg"
}
```

Load the image

```python
# load the image and show it
image = cv2.imread(args["image"])
plt_imshow("Original", image)
```

Rotate

```python
# 33-degrees counterclockwise
rotate33cc = imutils.rotate_bound(image, -33)
plt_imshow("Rotated Without Cropping", rotate33cc)
cv2.imwrite("../images/output/uss-freedom-rotate-33cc.jpg", rotate33cc)
# 45-degrees clockwise
```

```python
# 45-degrees clockwise
rotate45c = imutils.rotate_bound(image, 45)
plt_imshow("Rotated Without Cropping", rotate45c)
cv2.imwrite("../images/output/uss-freedom-rotate-45c.jpg", rotate45c)
```

```python
# 180-degrees clockwise
rotate180c = imutils.rotate_bound(image, 180)
plt_imshow("Rotated Without Cropping", rotate180c)
cv2.imwrite("../images/output/uss-freedom-rotate-180c.jpg", rotate180c)
```

```python
# 90-degrees counterclockwise
rotate90cc = imutils.rotate_bound(image, -90)
plt_imshow("Rotated Without Cropping", rotate90cc)
cv2.imwrite("../images/output/uss-freedom-rotate-90cc.jpg", rotate90cc)
```
