## Import Packages

```python
from matplotlib import pyplot as plt
import argparse
import cv2
import imutils
```

## Function to display images in Jupyter Notebooks

```python
# convert opencv BGR to RGB ordering
def plt_imshow(title, image):
    # convert image from BGR to RGB space and display
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.grid(False)
    plt.show()
```

## Pass the image to load

```python
args = {
    "image": "../images/output/uss-freedom-cropped.jpg"
}
```

## Load the image

```python
# load the original input image and display it to our screen
# loads the image set above from disk into image
image = cv2.imread(args["image"])
(h, w, c) = image.shape[:3]

# display the image dimensions
print("width: {} pixels".format(image.shape[1]))
print("heigth: {} pixels".format(image.shape[0]))
print("channels: {}".format(image.shape[2]))

plt_imshow("Image", image)
```

Resize without distortion and write to disk

```python
resized = imutils.resize(image, width=300)
cv2.imwrite("../images/output/uss-freedom-resized.jpg", resized)

print("width: {} pixels".format(resized.shape[1]))
print("heigth: {} pixels".format(resized.shape[0]))
print("channels: {}".format(resized.shape[2]))

plt_imshow("Resized via imutils", resized)
```
